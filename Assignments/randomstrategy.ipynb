```python
## Strategyless Game ##
# Computer is player 1, you are player 0 #

import numpy as np
import scipy.linalg as LA
rng = np.random.default_rng()

dim = 3 ## Use this to change dimension
A = [['X' for i in range(0,dim)]for j in range(0,dim)]
gofirst = 0     # use 1 for computer to go first, 0 for player first

## Chooses a random position for player1's move  
def p1_turn(A):
    turn_played = False
    while turn_played == False:
        i = rng.integers(0,dim)
        j = rng.integers(0,dim)
        if A[i][j]=="X":
            A[i][j]=1
            turn_played= True

## Takes input for player0's move
def p0_turn(A):
    p0_turn_played = False
    while p0_turn_played == False:
        p0_move = input("What is your move? ")
        p0_move = p0_move.split(",")
        i = int(p0_move[0])
        j = int(p0_move[1])
        if A[i-1][j-1]=="X":
            A[i-1][j-1]= 0
            p0_turn_played = True
        else:
            print("This spot is already taken.")
            
## Gameplay           
game_over = False
if gofirst ==1:
    for rep in range(0,(dim**2)//2):
        p1_turn(A)
        for row in A:
            for num in row:
                print(f"{num}",end=" ")
            print("")
        p0_turn(A)
        for row in A:
            for num in row:
                print(f"{num}",end=" ")
            print("")
        print("Player 1's Move:")
    if dim %2 ==1:
        p1_turn(A)
        for row in A:
            for num in row:
                print(f"{num}",end=" ")
            print("")
        print("")
                
    if int(LA.det(A)) == 0:
        print("Player 0 Wins!")
    else:
        print("Player 1 Wins!")
    
elif gofirst ==0:
    for rep in range(0,(dim**2)//2):
        p0_turn(A)
        for row in A:
            for num in row:
                print(f"{num}",end=" ")
            print("")
        print("Player 1's Move:")
        
        p1_turn(A)
        for row in A:
            for num in row:
                print(f"{num}",end=" ")
            print("")

        
    if dim%2 == 1:
        p0_turn(A)
        for row in A:
            for num in row:
                print(f"{num}",end=" ")
            print("")
        print("")
    
    if int(LA.det(A))==0:
        print("Player 0 Wins!")
    else:
        print("Player 1 Wins!")
```
