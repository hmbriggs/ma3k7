```python
###Â nxn game for n>3 ###
# computer plays 0, you play 1 #

import numpy as np
import scipy.linalg as LA
rng = np.random.default_rng()

dim = 5 ## Use this to change dimension
A = [['*' for i in range(0,dim)]for j in range(0,dim)]
gofirst = 0 


def p1move():
    global i
    global j
    p1_turn_played = False
    while p1_turn_played == False:
        p1_move = input("What is your move (in the form i,j)? ")
        p1_move = p1_move.split(",")
        i = int(p1_move[0])
        j = int(p1_move[1])
        if A[i-1][j-1]=="*":
            A[i-1][j-1]= 1
            p1_turn_played = True
        else:
            print("This spot is already taken.")

## Player 0 goes first ##
if gofirst ==0:
    full = False
    while full == False:
        if A[0][0] == '*':
            A[0][0] = 0
            for row in A:
                for num in row:
                    print(f"{num}",end=" ")
                print("")
        p1move()
        count = 0
        for row in A:
            for item in row:
                if item == '*':
                    count = count+1
        if count ==0:
            full = True
            if LA.det(A) == 0:
                print("det(A)= "+str(int(LA.det(A)))+" Player0 wins!")
            else:
                print("det(A)= "+str(int(LA.det(A)))+" Player1 wins!")
        for row in A:
            for num in row:
                print(f"{num}",end=" ")
            print("")
        if j == 1:
            if A[i-1][1] == '*':
                A[i-1][1] = 0
            else:
                unique = False
                while unique == False:
                    rand_i = rng.integers(0,dim)
                    rand_j = rng.integers(0,dim)
                    if A[rand_i][rand_j] =='*':
                        A[rand_i][rand_j] = 0
                        unique = True
        elif j == 2:
            if A[i-1][0] == '*':
                A[i-1][0] = 0
            else:
                unique = False
                while unique == False:
                    rand_i = rng.integers(0,dim)
                    rand_j = rng.integers(0,dim)
                    if A[rand_i][rand_j] =='*':
                        A[rand_i][rand_j] = 0
                        unique = True
        elif j == 3 :
            if A[i-1][3] == '*':
                A[i-1][3] = 0
            else:
                unique = False
                while unique == False:
                    rand_i = rng.integers(0,dim)
                    rand_j = rng.integers(0,dim)
                    if A[rand_i][rand_j] =='*':
                        A[rand_i][rand_j] = 0
                        unique = True
        elif j == 4:
            if A[i-1][2] == '*':
                A[i-1][2] = 0
            else:
                unique == False
                while unique == False:
                    rand_i = rng.integers(0,dim)
                    rand_j = rng.integers(0,dim)
                    if A[rand_i][rand_j] =='*':
                        A[rand_i][rand_j] = 0
                        unique = True
        else:
            unique = False
            while unique == False:
                counter = 0
                for k in range(0,dim):
                    for l in range(4,dim):
                        if A[k][l]=='*':
                            counter = counter +1
                if counter >0:
                    rand_i = rng.integers(0,dim)
                    rand_j = rng.integers(4,dim)
                    if A[rand_i][rand_j] =='*':
                        A[rand_i][rand_j] = 0
                        unique = True
                else:
                    rand_i = rng.integers(0,dim)
                    rand_j = rng.integers(0,dim)
                    if A[rand_i][rand_j] =='*':
                        A[rand_i][rand_j] = 0
                        unique = True
                        
        print("Player0's move:")            
        for row in A:
            for num in row:
                print(f"{num}",end=" ")
            print("")
            
        count = 0
        for row in A:
            for item in row:
                if item == '*':
                    count = count+1
        if count ==0:
            full = True
            if LA.det(A) == 0:
                print("det(A)= "+str(int(LA.det(A)))+" Player0 wins!")
            else:
                print("det(A)= "+str(int(LA.det(A)))+" Player1 wins!")
                

## Player1 goes first ##                
elif gofirst ==1:
    full = False
    for row in A:
            for num in row:
                print(f"{num}",end=" ")
            print("")
    while full == False:
        p1move()
        count = 0
        for row in A:
            for item in row:
                if item == '*':
                    count = count+1
        if count ==0:
            full = True
            if LA.det(A) == 0:
                print("det(A)= "+str(int(LA.det(A)))+" Player0 wins!")
            else:
                print("det(A)= "+str(int(LA.det(A)))+" Player1 wins!")
        for row in A:
            for num in row:
                print(f"{num}",end=" ")
            print("")
        if j == 1:
            if A[i-1][1] == '*':
                A[i-1][1] = 0
            else:
                unique = False
                while unique == False:
                    rand_i = rng.integers(0,dim)
                    rand_j = rng.integers(0,dim)
                    if A[rand_i][rand_j] =='*':
                        A[rand_i][rand_j] = 0
                        unique = True
        elif j == 2:
            if A[i-1][0] == '*':
                A[i-1][0] = 0
            else:
                unique = False
                while unique == False:
                    rand_i = rng.integers(0,dim)
                    rand_j = rng.integers(0,dim)
                    if A[rand_i][rand_j] =='*':
                        A[rand_i][rand_j] = 0
                        unique = True
        elif j == 3 :
            if A[i-1][3] == '*':
                A[i-1][3] = 0
            else:
                unique = False
                while unique == False:
                    rand_i = rng.integers(0,dim)
                    rand_j = rng.integers(0,dim)
                    if A[rand_i][rand_j] =='*':
                        A[rand_i][rand_j] = 0
                        unique = True
        elif j == 4:
            if A[i-1][2] == '*':
                A[i-1][2] = 0
            else:
                unique == False
                while unique == False:
                    rand_i = rng.integers(0,dim)
                    rand_j = rng.integers(0,dim)
                    if A[rand_i][rand_j] =='*':
                        A[rand_i][rand_j] = 0
                        unique = True
        else:
            unique = False
            while unique == False:
                counter = 0
                for k in range(0,dim):
                    for l in range(4,dim):
                        if A[k][l]=='*':
                            counter = counter +1
                if counter >0:
                    rand_i = rng.integers(0,dim)
                    rand_j = rng.integers(4,dim)
                    if A[rand_i][rand_j] =='*':
                        A[rand_i][rand_j] = 0
                        unique = True
                else:
                    rand_i = rng.integers(0,dim)
                    rand_j = rng.integers(0,dim)
                    if A[rand_i][rand_j] =='*':
                        A[rand_i][rand_j] = 0
                        unique = True
                        
        print("Player0's move:")            
        for row in A:
            for num in row:
                print(f"{num}",end=" ")
            print("")
            
        count = 0
        for row in A:
            for item in row:
                if item == '*':
                    count = count+1
        if count ==0:
            full = True
            if LA.det(A) == 0:
                print("det(A)= "+str(int(LA.det(A)))+" Player0 wins!")
            else:
                print("det(A)= "+str(int(LA.det(A)))+" Player1 wins!")
```
