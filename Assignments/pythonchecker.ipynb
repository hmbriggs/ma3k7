```python
import numpy as np
rng = np.random.default_rng()

n = 2
reps = 10

### Testing whether the grid has a 1-(n^2)-path ###
def path_test(grid, n):
    adjacency_counter = 0
    for i in range(0,n):
        for j in range(0,n):
            if grid[i][j] != n**2:  
                if i==0 and j ==0:
                    if grid[i+1][j]==grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                    if grid[i][j+1]==grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                elif i==0 and j==n-1:
                    if grid[i+1][j]==grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                    if grid[i][j-1]==grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                elif i==n-1 and j==0:
                    if grid[i-1][j]==grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                    if grid[i][j+1]==grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                elif i==n-1 and j==n-1:
                    if grid[i-1][j]==grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                    if grid[i][j-1]==grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1    
                elif i==0 and j!=0 and j!=n-1:
                    if grid[i+1][j]==grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                    if grid[i][j+1] ==grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                    if grid[i][j-1] == grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                elif j==0 and i!=0 and i!=n-1:
                    if grid[i+1][j]==grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                    if grid[i-1][j] ==grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                    if grid[i][j+1] == grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                elif i==n-1 and j!=0 and j!=n-1:
                    if grid[i-1][j]==grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                    if grid[i][j+1] ==grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                    if grid[i][j-1] == grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                elif j==n-1 and i!=0 and i!=n-1:
                    if grid[i+1][j]==grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                    if grid[i-1][j] ==grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                    if grid[i][j-1] == grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                else:
                    if grid[i+1][j]==grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                    if grid[i-1][j] ==grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                    if grid[i][j+1] == grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
                    if grid[i][j-1] == grid[i][j]+1:
                        adjacency_counter = adjacency_counter+1
    return adjacency_counter
            
###################################################

grid_counter = 0

for k in range(0,reps):

    ### Setting up the grid ###
    grid = [['*' for i in range(0,n)]for j in range(0,n)]

    table_pos = []
    for i in range(0,n):
        for j in range(0,n):
            table_pos.append([i,j])

    count = 0
    for i in range(0,len(table_pos)):
        count = count+1
        index = table_pos[rng.integers(0,len(table_pos))]
        table_pos.remove(index)

        grid[index[0]][index[1]] = count


    ### Checking whether the grid has a path ###

    if path_test(grid,n) == (n**2)-1:
        grid_counter = grid_counter+1
        
        for row in grid:
            print("\033[34m"+f"{row}"+"\033[0m")
    else:
        for row in grid:
            print(row)
    print("")
```
