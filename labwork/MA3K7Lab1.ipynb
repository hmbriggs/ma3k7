```python
import numpy as np
import matplotlib.pyplot as plt
```

### 2 Logarithmic Plots

```python
x = np.linspace(0,100,20)
y = x**8
y2 = x**9

plt.plot(x,y,marker="o", markersize=3, color = "blue",label="$y=x^8$")
plt.plot(x,y2, marker="o", markersize=3, color = "red",label = "$y=x^9$")
plt.title("Using plt.plot")
plt.legend(['$y=x^8$','$y=x^9$'])
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(1,100)
plt.grid('on')
plt.show()
```

```python
x = np.logspace(0,2,20) ##Better when using logs as makes points evenly spaced
y = x**8
y2 = x**9

plt.loglog(x,y,"-ko",x,y2,"-rx",ms=5)
plt.title("Using plt.loglog")
plt.show()

plt.semilogx(x,y,x,y2,marker="o", markersize=3)
plt.title("Using plt.semilogx")
plt.show()

plt.semilogy(x,y,x,y2,marker="o", markersize=3)
plt.title("Using plt.semilogy")
plt.show()
```
