```python
from turtle import Screen, Turtle
from random import choice, randint

### Adds particles to the box ###
def addball():
    part = Turtle('circle')
    part.shapesize(part_diameter / cursor_size)
    part.color(choice(colours))
    part.penup()
    part.speed('fastest')
    x, y = randint(part_diameter - width/2, width/2 - part_diameter), randint(part_diameter - width/2, width/2 - part_diameter)
    part.goto(x, y)
    part.dy = randint(-8, 8)
    part.dx = randint(-8, 8)

    particles.append(part)

    num_parts = len(particles)

### Checks if two particles collide and changes species accordingly ###
def collision(other_part):
    for part in particles:
        if part != other_part and part.distance(other_part) < part_diameter:
            other_part.dy *=-1
            other_part.dx *= -1
            part.dy *= -1
            part.dx *= -1
            if part.color() != other_part.color():
                if part.color() == ('red', 'red'):
                    if other_part.color() == ('blue', 'blue'):
                        part.color("green")
                        other_part.color("green")
                    elif other_part.color() == ('green', 'green'):
                        part.color("blue")
                        other_part.color("blue")
                elif part.color() == ('blue', 'blue'):
                    if other_part.color() == ('green', 'green'):
                        part.color("red")
                        other_part.color("red")
                    elif other_part.color() == ('red', 'red'):
                        part.color("green")
                        other_part.color("green")
                elif part.color() == ('green', 'green'):
                    if other_part.color() == ('blue', 'blue'):
                        part.color("red")
                        other_part.color("red")
                    elif other_part.color() == ('red', 'red'):
                        part.color("blue")
                        other_part.color("blue")
                        
def tick():
    for part in particles:
        part.sety(part.ycor() + part.dy)
        part.setx(part.xcor() + part.dx)
        if part.ycor() > height/2-part_diameter or part.ycor() < -height/2+part_diameter:
            part.dy *= -1
        if part.xcor() > width/2-part_diameter or part.xcor() < -width/2+part_diameter:
            part.dx *= -1
        collision(part)

    screen.update()
    screen.ontimer(tick, 60)

    
part_diameter = 30
width, height = 400, 400
colours = ['red', 'green', 'blue']
no_part = 5
cursor_size = 20

screen = Screen()
screen.setup(width, height)
screen.tracer(False)

particles = []

for i in range(no_part):
    addball()

tick()
screen.mainloop()
```
